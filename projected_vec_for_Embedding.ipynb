```python
import numpy as np
import os
```

```python
data = np.load('../15D_atten/0.01_0.01_0.001_0.01_Edims200.npz')
```

```python
for file in data.files:
    print(file,data[file].shape)
```

```python
def project(project_vec,be_projected_vec):
    #vector projection for x onto y 
    x = data['E'][project_vec]
    y = data['E'][be_projected_vec]
    after_px = y * np.dot(x, y) / np.dot(y, y)
    return after_px
```

```python
for j in range(1):
    for i in range(200):
        print('Project',i,'dimension onto', j, 'dimension')
        print(j,'dimension',data['E'][j])
        print('after project',project(i,j))
        print('----------------------------------------------------')
```
